### Test 1: Layerwise Initialization, retrain Inner Layers

```python
import torch, os
from timm.models import create_model
import resmlp_model_tight

model = create_model(
        "resmlp_tight_24",
        pretrained=True,
        num_classes=1000,
        drop_rate=0.0,
        drop_path_rate=0.1,
        drop_block_rate=None,
        img_size=224
    )
```

```python
import torchsummary
torchsummary.summary(model, (3, 224, 224), device="cpu")
```

```python
in_features = model.blocks[0].inner.in_features
model.blocks[0].inner = torch.nn.Linear(in_features, in_features, bias=True)
model.blocks[0].inner.weight = torch.nn.Parameter(torch.eye(in_features))
```
